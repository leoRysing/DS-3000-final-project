```python
tuition_by_incom_0To30000 = tuition_by_incom[tuition_by_incom["income_lvl"] == "0 to 30,000"]
tuition_by_incom_0To30000 = tuition_by_incom_0To30000.drop(labels = ["state", "income_lvl"], axis = 1)
new_Names_block1 = {"net_cost":"net_cost_0_to_30"}
tuition_by_incom_0To30000 = tuition_by_incom_0To30000.rename(columns=new_Names_block1)
tuition_by_incom_0To30000
```

```python
tuition_by_incom_300001 = tuition_by_incom[tuition_by_incom["income_lvl"] == "30,001 to 48,000"]
tuition_by_incom_300001 = tuition_by_incom_300001.drop(labels = ["state", "income_lvl"], axis = 1)
new_Names_block2 = {"net_cost":"net_cost_30_to_48"}
tuition_by_incom_300001 = tuition_by_incom_300001.rename(columns=new_Names_block2)
tuition_by_incom_300001.head(15)
```

```python
tuition_by_incom_48001 = tuition_by_incom[tuition_by_incom["income_lvl"] == "48_001 to 75,000"]
tuition_by_incom_48001 = tuition_by_incom_48001.drop(labels = ["state", "income_lvl"], axis = 1)
new_Names_block3 = {"net_cost":"net_cost_48_to_75"}
tuition_by_incom_48001 = tuition_by_incom_48001.rename(columns=new_Names_block3)
tuition_by_incom_48001
```

```python
tuition_by_incom_75001 = tuition_by_incom[tuition_by_incom["income_lvl"] == "75,001 to 110,000"]
tuition_by_incom_75001 = tuition_by_incom_75001.drop(labels = ["state", "income_lvl"], axis = 1)
new_Names_block4 = {"net_cost":"net_cost_75_to_110"}
tuition_by_incom_75001 = tuition_by_incom_75001.rename(columns=new_Names_block4)
tuition_by_incom_75001
```

```python
tuition_by_incom_111000 = tuition_by_incom[tuition_by_incom["income_lvl"] == "Over 110,000"]
tuition_by_incom_111000 = tuition_by_incom_111000.drop(labels = ["state", "income_lvl"], axis = 1)
new_Names_block5 = {"net_cost":"net_cost_over_110"}
tuition_by_incom_111000 = tuition_by_incom_111000.rename(columns=new_Names_block5)
tuition_by_incom_111000
```

```python
from functools import reduce
```

```python
data_frames = [tuition_by_incom_0To30000, tuition_by_incom_300001, 
              tuition_by_incom_48001, tuition_by_incom_75001, tuition_by_incom_111000]


tuit_inc_merged = reduce(lambda  left,right: pd.merge(left,right,
                                                      on=["name", "year", "campus", "total_price"],
                                                      how='outer'), data_frames)
#tuit_inc_merged = pd.merge(tuition_by_incom_0To30000, 
                       #    tuition_by_incom_300001, on=["name", "year", "campus", "total_price"],
                        #  how="inner")
tuit_inc_merged
```

```python
tuit_inc_merged = tuit_inc_merged.dropna()
tuit_inc_merged = tuit_inc_merged.drop_duplicates()
tuit_inc_merged
```

```python
#tuit_inc_merged
tuit_inc_merged.reset_index()
tuit_inc_merged.head(25)
```

```python
tuit_inc_merged = tuit_inc_merged.reset_index()
```

```python
sliced_diversity = diversity[['name', 'total_enrollment', 'category', 'enrollment']].where((diversity['category'] == 'White') 
                                                    | (diversity['category'] == 'Asian')
                                                   | (diversity['category'] == 'Black')
                                                   | (diversity['category'] == 'Hispanic')
                                                   | (diversity['category'] == 'American Indian / Alaska Native')).dropna()
sliced_diversity
```

```python
diversity_asian = sliced_diversity[sliced_diversity["category"] == "Asian"]
diversity_asian = diversity_asian.drop(labels = ["category"], axis = 1)
new_Names_block1 = {"enrollment":"asian_enrollment"}
diversity_asian = diversity_asian.rename(columns=new_Names_block1)
diversity_asian
```

```python
diversity_white = sliced_diversity[sliced_diversity["category"] == "White"]
diversity_white = diversity_white.drop(labels = ["category"], axis = 1)
new_Names_block1 = {"enrollment":"white_enrollment"}
diversity_white = diversity_white.rename(columns=new_Names_block1)
diversity_white
```

```python
diversity_black = sliced_diversity[sliced_diversity["category"] == "Black"]
diversity_black = diversity_black.drop(labels = ["category"], axis = 1)
new_Names_block1 = {"enrollment":"black_enrollment"}
diversity_black = diversity_black.rename(columns=new_Names_block1)
diversity_black
```

```python
diversity_hispanic = sliced_diversity[sliced_diversity["category"] == "Hispanic"]
diversity_hispanic = diversity_hispanic.drop(labels = ["category"], axis = 1)
new_Names_block1 = {"enrollment":"hispanic_enrollment"}
diversity_hispanic = diversity_hispanic.rename(columns=new_Names_block1)
diversity_hispanic
```

```python
diversity_native = sliced_diversity[sliced_diversity["category"] == "American Indian / Alaska Native"]
diversity_native = diversity_native.drop(labels = ["category"], axis = 1)
new_Names_block1 = {"enrollment":"american_indian_alaska_native_enrollment"}
diversity_native = diversity_native.rename(columns=new_Names_block1)
diversity_native
```

```python
df_diversity = [diversity_asian, diversity_white, diversity_black, diversity_native, diversity_hispanic]


div_merged = reduce(lambda  left,right: pd.merge(left,right,
                                                      on=["name", 'total_enrollment'],
                                                      how='outer'), df_diversity)
#tuit_inc_merged = pd.merge(tuition_by_incom_0To30000, 
                       #    tuition_by_incom_300001, on=["name", "year", "campus", "total_price"],
                        #  how="inner")
div_merged
```

```python
div_merged.to_csv("raw/diversity_merged.csv")
```
