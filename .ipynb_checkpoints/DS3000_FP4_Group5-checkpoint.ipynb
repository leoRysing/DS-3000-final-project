## 1. INTRODUCTION

 

<h4>Problem Statement</h4>
 
For our project, we wanted to tackle the issue of college tuition being a possible hindrance for diversity on campus, especially for groups who on average fall more closely in lower income brackets. As university students, we wanted to do this project because we have felt that tuition costs have been growing tremendously in the past years, and we wanted to see the actual effects this raised tuition has on university demographics. In effect, our topic is an analysis of student tuition by income bracket and how it affects the student body population.
 
Our first objective is primarily to explore the relationship between different college tuitions, university types, and total student enrollment influences demographic breakdowns of students. The second objective is to see and note which range of tuition blocks, if any, provide a much more balanced student body.
 
<h4>Significance of the Problem</h4>
 
It is important to tackle this, because college is often seen as a requirement step for many jobs today,
but this doesn’t mean that everybody is able to make use of this incredibly important resource. Because of the rapidly inflating tuition, we want to see if tuition are still reasonable for the majority of Americans, as well as to predict how
movement in diverse student bodies impacts this as well.
 
We want to see if increased debt from increased costs might be hampering certain factors that decide
who can go to college. It might also be useful to argue in favor of adding more control towards college
tuition. Depending on the findings we get, it might provide evidence for student activists and universities that claim to wish to foster diversity on campus to control tuition. At the very least, it might help keep eyes on this issue, and add pressure for administrations when they decide on tuition costs.
 
<h4>Questions</h4>
 
Based on the importance of the issue of rising college tuition and diversity on college campuses, we hope to answer the following questions through analyzing out data:
- Which factors (income breakdown, location, price, in-state and out-of-state enrollment, etc) have an impact on racial diversity on college campuses?
- Does the cost of a college or university play into how diverse its campus will be?
- Does the race of a student impact the college they choose to attend?
- Do different groups get impacted by the features the same way?
- Which regression model best illustrates the relationship between different attributes of a university and its diversity?

- Hypothesis about features:
   - Null: Various attributes of colleges and universities, such as income, cost, location, and degrees offered, will have no impact on the racial diversity of universities. In addition, race will have no impact on the type of college a student will attend.

### 2.1. Data Acquisition
 
#### The Original Source of the Data
The data we used was acquired from a kaggle dataset created by Jesse Mostipak. Much of the data originates from the US Department of Education, and Ms. Mostipak cites their own code that was used to clean the aggregated data. The last time this dataset was updated was approximately a year ago. The data was downloaded from kaggle and added to our Git Repository. The datasets are collectively about college tuition, costs, enrollment, and demographics.
 
#### Feature Description
The datasets, based off the US Department of Education's data, is based on data from the 2016-2019 school years. Not all data within these datasets are quantitative, but can be encoded as such as shown in later analysis. There are disparities in unique values between the diversity data set and tuition dataset, indicating that non-existent values will appear after merging and need to be cleaned. These features are our independent variables.
 
Features and their descriptions (original count: 17 | addition of features: 21):
* total_enrollment : The number of students enrolled at the institution in the 2016-2019 school years
* total_price : The average tuition cost compounded with other additional costs such as housing and dining
* year : The year the data was recorded, ranging from 2016 to 2019
* on_campus : OneHotEncoded data that indicates whether this cost is the price of being on campus
* off_campus : OneHotEncoded data that indicates whether this cost is the price of being off campus, means that room and board should not be accounted for
* for_profit : OneHotEncoded data that indicates whether this institution is for profit
* private : OneHotEncoded data that indicates whether this institution is private
* public : OneHotEncoded data that indicates whether this institution is for public
* two_year : OneHotEncoded data that indicates whether this is a two year institution, which offers associates degrees
* four_year : OneHotEncoded data that indicates whether this is a four year institution, which offers bachelor's degrees
* net_cost_0_to_30 : The net cost of attending the institution for students whose families fall into the 0-30k income bracket
* net_cost_30_to_48 : The net cost of attending the institution for students whose families fall into the 30k-48k income bracket
* net_cost_48_to_75 : The net cost of attending the institution for students whose families fall into the 48k-75k income bracket
* net_cost_75_to_110 : The net cost of attending the institution for students whose families fall into the 75k-110k income bracket
* net_cost_over_110 : The net cost of attending the institution for students whose families fall into the over 110k income bracket
* state_code : A numerical representation of state abbreviations. The state can be found using a dictionary in the data wrangling. The numbers range from 0-49 for the 50 states
* room_and_board : the non-tuition costs requested by the university for living on campus
* in_state_tuition : The cost of tuition for students that live in the same state as that institution
* out_of_state_tuition : The cost tuition the institution for students that do not live in the same state as that institution
* in_state_total : The total cost of attending the institution for students that live in the same state as that institution
* out_of_state_tuition : The total cost of attending the institution for students that live in the same state as that institution
 
Targets and their descriptions (count: 5): These are the dependent variables
* american_indian_alaska_native_enrollment : The number of students of Native American or Alaskan descent enrolled at the institution
* asian_enrollment : The number of students of Asian descent enrolled at the institution
* black_enrollment : The number of students of African descent enrolled at the institution
* hispanic_enrollment : The number of students of Hispanic descent enrolled at the institution
* white_enrollment : The number of students of Caucasian descent enrolled at the institution

### 2.2. Data Analysis
 
#### Predictive Model
We are going to be predicting the expected student body size for 5 specific demographics (White students, African American students, Hispanic students, Indigenous students, and Asian-American students). In  other words, given the data about various school tuition prices for different income brackets, school type, on campus status, total enrollment, and state location, we will be predicting the expected number of students at that school of each of the selected demographic bodies. For our prediction, we will use the above collection of features to train our model (after some slight standardization). We think these are important predictors, because if we can reliably predict them, then we can play around to see which combination of tuition costs can maximize diversity on campus.
 
#### A Supervised Learning Problem
This is a supervised learning problem, since features are named, and we are not looking to cluster results. It is subcategorized to a regression problem - we are seeking to obtain a/some continuous output variable(s).
 
#### Machine Learning Algorithms To Be Applied
We will be using Lasso, KNN, and Ridge regression models. We are however somewhat sceptical of linear fit in our model; this is why we have chosen to include kNN regression. We chose these three, because Lasso and Ridge are fairly robust linear models, and kNN is the main algorithm we could think of that didn't directly regress using linear regression. We originally looked at polynomial regression and SVM regression, but realized that both of those did not really fit our data.

Here we are making a dictionary of all of the regression models we are going to use, as well as a dictionary of all of the racial groups we will be analyzing.

```python
#creating a dictionary of estimators
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge


estimators = {"Lasso Regression" : Lasso(tol=0.1, alpha=0.001),
              "kNN" : KNeighborsRegressor(),
              "Ridge Regression" : Ridge(alpha = 100)}
              
inc_div_scaled_features = inc_div_scaled.iloc[:,0:20]

#creating a dictionary of ethnic groups
ethnic_group = {"Asian" : (inc_div_scaled_features, target1), 
                "Black" : (inc_div_scaled_features, target3), 
                "Hispanic" : (inc_div_scaled_features, target5), 
                "Native American/Alaskan" : (inc_div_scaled_features, target4), 
                "White" : (inc_div_scaled_features, target2)}
```

Splitting the data into training and testing sets, which will be used for each feature.

```python
from sklearn.model_selection import train_test_split

def split_train_test(features, target) :
    split = train_test_split(features, target, random_state = 3000)
    return split
```

For each ethnic group, add to a dictionary with its own training and testing set that can be used for initial application of regressors and after tuning.

```python
def target_ML_sets() :
    ethnic_ML_sets = {}
    for ethnicity, dataset in ethnic_group.items() :
        feature_set, target_set = dataset
        ethnic_ML_sets[ethnicity] = split_train_test(feature_set, target_set)
    return ethnic_ML_sets

ethnic_ML_sets = target_ML_sets()
```

Selecting the best 12 features with K-Best Selection.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

def select_features() :
    selected = []
    for ethnicity, train_test_sets in ethnic_ML_sets.items() :
        X_train, X_test, y_train, y_test = train_test_sets
        
        #feature selection
        select = SelectKBest(score_func = f_regression, k=12)
        select.fit(X_train, y_train)
        #transform sets
        X_train_selected = select.transform(X_train)
        X_test_selected = select.transform(X_test)
        selected.append(select)
    return selected
```

```python
selected = select_features()
```

### 3.3. Model Training

#### Creating the Models

The two function defined below run Lasso, Ridge, and KNeighbors regression models on all 5 target variables and their associated features. apply_algorithms() uses all features while apply_algorithms_selected() uses K-Best variable selection to create training and testing sets.

```python
#function applies regressors
#evaluates performance using r^2 values and mean-squared-error values 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

def apply_algorithms() :
    for ethnicity, train_test_sets in ethnic_ML_sets.items() :
        print("\tLooking at ", ethnicity, " target group:", sep="")
        print("\t–––––––––––––––––––––––––––––––––––––––")
        X_train, X_test, y_train, y_test = train_test_sets
        for estimator, algorithm in estimators.items() :
            model = algorithm.fit(X=X_train, y=y_train)
        
            print("\t", estimator, ":", sep="")
            print("\t\t", "R-squared value for training set:  ", r2_score(y_train, model.predict(X_train)), sep="")
            print("\t\t", "R-squared value for testing set:  ", r2_score(y_test, model.predict(X_test)), sep="")
            print("\t\t", "MSE value for training set:  ", mean_squared_error(y_train, model.predict(X_train)), sep="")
            print("\t\t", "MSE value for testing set:  ", mean_squared_error(y_test, model.predict(X_test)), sep="")
            print("\n")
        print("\t========================================")
```

```python
#Feature selection using K-Best Selection: want to retain the best 12 features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

def apply_algorithms_selected() :
    for ethnicity, train_test_sets in ethnic_ML_sets.items() :
        print("\tLooking at ", ethnicity, " target group:", sep="")
        print("\t–––––––––––––––––––––––––––––––––––––––")
        X_train, X_test, y_train, y_test = train_test_sets
        
        #feature selection
        select = SelectKBest(score_func = f_regression, k=12)
        select.fit(X_train, y_train)
        #transform sets
        X_train_selected = select.transform(X_train)
        X_test_selected = select.transform(X_test)
        selected.append(select)
        #the selected features
        print("\tSelected features after RFE:")
        for i in range(len(inc_div_scaled_features.columns)):
            if select.get_support()[i] == True:
                print("\t", inc_div_scaled_features.columns[i])
        print("\n", end="")
        for estimator, algorithm in estimators.items() :
            model = algorithm.fit(X=X_train_selected, y=y_train)
        
            print("\t", estimator, ":", sep="")
            print("\t\t", "R-squared value for training set:  ", r2_score(y_train, model.predict(X_train_selected)), sep="")
            print("\t\t", "R-squared value for testing set:  ", r2_score(y_test, model.predict(X_test_selected)), sep="")
            print("\t\t", "MSE value for training set:  ", mean_squared_error(y_train, model.predict(X_train_selected)), sep="")
            print("\t\t", "MSE value for testing set:  ", mean_squared_error(y_test, model.predict(X_test_selected)), sep="")
            print("\n")
        print("\t========================================")
```

### 3.4. Model Optimization

The models above have the possibility of improving in performance. They also run the risk of overfitting the data. Thus, we tune the models to either reduce/increase model complexity (in the case of Lasso and Ridge) or discover the optimal amount of outside influence (KNeighbors).

```python
#Tuning parameters
param_grid_ridge = {"alpha" : [0.001, 0.01, 0.1, 1, 10, 100], 
              "solver" : ["auto", "svd", "cholesky", "lsqr", "sparse_cg", "sag", "saga"]}
param_grid_lasso = {"alpha" : [0.001, 0.01, 0.1, 1, 10, 100]}
param_grid_kNN = {"metric" : ["euclidean", "manhattan", "minkowski"], "n_neighbors" : [3, 5, 7, 11]}
```

```python
#GridSearch for Ridge Method
#import GridSearchCV and fit GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV


def hypertune_ridge() :
    best_params = {}
    grid_search = GridSearchCV(Ridge(), param_grid_ridge, cv=5)
    
    for ethnicity, train_test_sets in ethnic_ML_sets.items() :
        print("\tLooking at ", ethnicity, " target group:", sep="")
        print("\t–––––––––––––––––––––––––––––––––––––––")
        X_train, X_test, y_train, y_test = train_test_sets
        
        #feature selection
        select = SelectKBest(score_func = f_regression, k=12)
        select.fit(X_train, y_train)
        #transform sets
        X_train_selected = select.transform(X_train)
        X_test_selected = select.transform(X_test)
        
        grid_search.fit(X=X_train_selected, y=y_train)
        
        print("\t\tBest parameters: ", grid_search.best_params_)
        print("\t\tBest cross-validation score: ", grid_search.best_score_)
        print("\t\tTest set score: ", grid_search.score(X_test_selected, y_test))
        best_params[ethnicity] = (grid_search.best_params_, X_train_selected, X_test_selected, y_train, y_test)
    return best_params
```

```python
#GridSearch for Lasso Method
#import GridSearchCV and fit GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

def hypertune_lasso() :
    best_params = {}
    grid_search = GridSearchCV(Lasso(tol=0.1), param_grid_lasso, cv=5)
    
    for ethnicity, train_test_sets in ethnic_ML_sets.items() :
        print("\tLooking at ", ethnicity, " target group:", sep="")
        print("\t–––––––––––––––––––––––––––––––––––––––")
        X_train, X_test, y_train, y_test = train_test_sets
        
        #feature selection
        select = SelectKBest(score_func = f_regression, k=12)
        select.fit(X_train, y_train)
        #transform sets
        X_train_selected = select.transform(X_train)
        X_test_selected = select.transform(X_test)
        
        grid_search.fit(X=X_train_selected, y=y_train)
        
        print("\t\tBest parameters: ", grid_search.best_params_)
        print("\t\tBest cross-validation score: ", grid_search.best_score_)
        print("\t\tTest set score: ", grid_search.score(X_test_selected, y_test))
        best_params[ethnicity] = (grid_search.best_params_, X_train_selected, X_test_selected, y_train, y_test)
    return best_params
```

```python
#GridSearch for kNN Method
#import GridSearchCV and fit GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

def hypertune_kNN() :
    best_params = {}
    grid_search = GridSearchCV(KNeighborsRegressor(), param_grid_kNN, cv=5)
    
    for ethnicity, train_test_sets in ethnic_ML_sets.items() :
        print("\tLooking at ", ethnicity, " target group:", sep="")
        print("\t–––––––––––––––––––––––––––––––––––––––")
        X_train, X_test, y_train, y_test = train_test_sets
        
        #feature selection
        select = SelectKBest(score_func = f_regression, k=12)
        select.fit(X_train, y_train)
        #transform sets
        X_train_selected = select.transform(X_train)
        X_test_selected = select.transform(X_test)
        
        grid_search.fit(X=X_train_selected, y=y_train)
        
        print("\t\tBest parameters: ", grid_search.best_params_)
        print("\t\tBest cross-validation score: ", grid_search.best_score_)
        print("\t\tTest set score: ", grid_search.score(X_test_selected, y_test))
        best_params[ethnicity] = (grid_search.best_params_, X_train_selected, X_test_selected, y_train, y_test)
    return best_params
```

Display hyperparameter tuning results:

```python
print("MACHINE LEARNING MODELS AFTER HYPERPARAMETER TUNING\n")

print("Ridge Regression with Tuning:\n")
best_ridge = hypertune_ridge()
print("\n", end="")
print("=========================================")
print("Lasso Regression with Tuning:\n")
best_lasso = hypertune_lasso()
print("\n", end="")
print("=========================================")
print("kNN Regression with Tuning:\n")
best_knn = hypertune_kNN()
print("\n", end="")
```

### Research questions
 
Can you determine a university's demographic breakdown just by looking at its costs? For some demographic more than others
 
***Which factors (income breakdown, location, price, in-state and out-of-state enrollment, etc) have an impact on racial diversity on college campuses?***
 
We found that different target demographics were impacted by different features. We found that each instance of feature selection had the private and public university features, alongside the net cost for the bottom 3 income brackets, and total price. Some models had other income brackets, two target groups (Asian American and white) also selected state as an important feature. Room and board cost, in-state tuition, and out-of-state tuition were other common selected features, which some targets felt were important, and others didn't.
 
We also found that there were two main visual patterns for our groups. As noted above, white and asian american enrollment seemed tied to very similar features, while African American, Indigenous, and Hispanic enrollment more closely matched with each other.
 
***Does the cost of a college or university play into how diverse its campus will be?***
 
This seems to currently be a tentative maybe, or a yes with limitations. All of the models selected total price and some form of net cost and other financial breakdown as being important to predicting diversity, as defined in the data analysis section (diversity as number of students of various demographics). However, from our visualizations, we can see that there isn't a strong negative linear correlation between the two. As of now it seems plausible that there is some connection between higher prices tetering diversity, but we need to apply more metrics to these combinations of feature and combo before we answer this question.
 
***Does the race of a student impact the college they choose to attend?***
 
Our data analysis did not specifically bring about an answer to this question. What we know is that each target group had different graphical patterns in our scatterplot, showing that public and private schools pulled students of different racial and ethnic backgrounds to differing degrees.
 
***Do different groups get impacted by the features the same ways?***
 
This is a definitive yes. As we saw from our feature selection, different features/ factors were selected as statistically significant for our models between the different target sets. No target group had exactly the same collection of most important features. On top of this, the additional takeaway can be made that although all the groups do not share a uniform impact with the features, there were certain overall patterns for which features different target groups followed (White and asian american target models showed this in their collection of 2-3 variables witch close to no linear correlation, one of which was the state of the university).
 
***Which regression model best illustrates the relationship between different attributes of a university and its diversity?***
 
As we predicted in our hypothesis, the kNN regression model illustrated the relationship between the attributes and university diversity. It was the only training model that was able to break through a 50% accuracy for our regression, and the only one that could account for more than 60% of the variation of our data (Linear and Ridge each accounted for under 20%).
 
### Reflecting on Our Findings

Tying back to our original signifcance, we wanted to see impacts of tuition information on diversity on campus, to see if raising tuitions are hurting efforts to overcome systematic differences different groups in America face. 

Based on the results we got, we cannot strongly conclude a linear negative correlation between higher tuition and decreased representation on campuses. It appears to have at most a weak, small, linear correlation in some places, but the data shows a very non linear form. 
 
### Conclusion
Our main conclusion, as outlined above, is that while it seems to be possible to predict how many students of a target group will enroll in a school based on the features we selected, a non linear model would be likely able to communicate more information and much more closely fit our data. We found slight evidence that expensive costs somewhat deter diversity (as defined by proportion of students of a target group present at a university), but we would need to here as well look at it with a fit model different from linear regression. We unfortunately were not able to strongly answer our question about race impacting what university a student would attend, since our data model did not use race as a feature. 
 
For the future, the main things we wanted to do was look at data from more years potentially, to actually get an analysis of effects of rising costs over the years. We also wanted to connect our data with data showing state demographic data, to see which type of school (public or private) is more representative of a given state. The biggest major extension we would want to do is find more non linear regression models we could apply to fit our data with, to actually compare different models that would be able to somewhat fit our data.
